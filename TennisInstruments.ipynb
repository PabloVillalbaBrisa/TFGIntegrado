```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications import imagenet_utils
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras import backend as K
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imageio as io
import gc

def predict_image(model_name, image_source):
    # Importar el modelo
    MODELS = {
        "vgg16": VGG16
    }

    #Tamaño entrada VGG16 y función preprocesamiento de imagen
    input_shape = (224, 224)
    preprocess = imagenet_utils.preprocess_input

    # Carga modelo
    print("[INFO]: Cargando {}...".format(model_name))
    Network = MODELS[model_name]
    model = Network(weights='imagenet')

    # Carga imagen
    print("[INFO]: Cargando y pre-procesando la imagen...")
    image = load_img(image_source, target_size=input_shape)
    image = img_to_array(image)

    # Extensión dimensiones
    image = np.expand_dims(image, axis=0)

    # Preprocesamiento de imagen
    image = preprocess(image)

    print("[INFO] Clasificando la imagen con '{}'...".format(model_name))
    predictions = model.predict(image)

    # Obtener etiquetas predicciones
    labels = imagenet_utils.decode_predictions(predictions)

    competitive_facts = False
    tennis_classes = ["tennis ball", "racket", "scoreboard"]
    for (_, label, _) in labels[0]:
        if any(tennis_class in label.lower() for tennis_class in tennis_classes):
            competitive_facts = True
            break

    # Clasificar según instrumentos de tenis
    classification = "with tennis instruments" if competitive_facts else "without tennis instruments"
    # Mostrar clasificación
    print("La imagen es clasificada como: {}".format(classification))

    # Liberar memoria
    K.clear_session()
    gc.collect()


    # Devolver la clasificación en lugar del modelo para evitar problemas de memoria
    return classification
```

```python
import os
import pandas as pd
import gc

directorio_raiz = '/content/drive/My Drive/TFG PABLO VILLALBA BRISA/TFG_Dataset'
csv_path = '/content/drive/My Drive/TFG PABLO VILLALBA BRISA/Archivos/insta_posts_kai.csv'

def extraer_shortcode(nombre_imagen):
    if nombre_imagen.endswith("UTC.jpg"):
        return nombre_imagen.replace(".jpg", "")
    elif nombre_imagen.endswith("UTC_1.jpg"):
        return nombre_imagen.replace("_1.jpg", "")
    return None

predicciones = {'post': [], 'Clase': []}

batch_size = 10
image_paths = []
shortcodes = []

for subdir, _, files in os.walk(directorio_raiz):
    for filename in files:
        if filename.endswith("UTC.jpg") or filename.endswith("UTC_1.jpg"):
            shortcode = extraer_shortcode(filename)
            if not shortcode:
                continue

            image_path = os.path.join(subdir, filename)
            image_paths.append(image_path)
            shortcodes.append(shortcode)

            if len(image_paths) == batch_size:

                preds_batch = []
                for img_path in image_paths:
                    pred = predict_image("vgg16", img_path)
                    preds_batch.append(pred)

                predicciones['post'].extend(shortcodes)
                predicciones['Type'].extend(preds_batch)

                image_paths = []
                shortcodes = []

                del preds_batch
                gc.collect()


#si faltan imágenes
if image_paths:
    preds_batch = []
    for img_path in image_paths:
        pred = predict_image("vgg16", img_path)
        preds_batch.append(pred)

    predicciones['post'].extend(shortcodes)
    predicciones['Clase'].extend(preds_batch)

    del preds_batch
    gc.collect()

df_predicciones = pd.DataFrame(predicciones)

df = pd.read_csv(csv_path)

df = pd.merge(df, df_predicciones, on='post', how='left')

df.to_csv(csv_path, index=False)
```

```python
import os
import pandas as pd
import gc

#Directorio raíz que contiene los subdirectorios con imágenes
directorio_raiz = '/content/drive/My Drive/TFG PABLO VILLALBA BRISA/TFG_Dataset'
#Archivo csv destino
nuevo_csv_path = '/content/drive/My Drive/TFG PABLO VILLALBA BRISA/Archivos/instrument_predictions.csv'

#Función para extraer el shortcode del nombre de la imagen
def extraer_shortcode(nombre_imagen):
    if nombre_imagen.endswith("UTC.jpg"):
        return nombre_imagen.replace("UTC.jpg", "")
    elif nombre_imagen.endswith("UTC_1.jpg"):
        return nombre_imagen.replace("UTC_1.jpg", "")
    return None

#Diccionario para almacenar las predicciones
predicciones = {'username': [], 'post': [], 'type': []}

#Tamaño del lote
batch_size = 10
image_paths = []
shortcodes = []
usernames = []

#Iterar sobre los subdirectorios en el directorio raíz
for subdir, _, files in os.walk(directorio_raiz):
    for filename in files:
        if filename.endswith("UTC.jpg") or filename.endswith("UTC_1.jpg"):
            #Extraer el shortcode
            shortcode = extraer_shortcode(filename)
            if not shortcode:
                continue

            #Ruta completa de la imagen
            image_path = os.path.join(subdir, filename)
            image_paths.append(image_path)
            shortcodes.append(shortcode)

            #Suponiendo que el nombre del usuario esté en el subdirectorio
            usuario = os.path.basename(subdir)
            usernames.append(usuario)

            # Si el tamaño del lote alcanza el límite o es la última imagen
            if len(image_paths) == batch_size:
                # Obtener las predicciones del modelo para el lote de imágenes
                preds_batch = []
                for img_path in image_paths:
                    pred = predict_image("vgg16", img_path)
                    preds_batch.append(pred)

                # Agregar las predicciones al diccionario
                predicciones['username'].extend(usernames)
                predicciones['post'].extend(shortcodes)
                predicciones['type'].extend(preds_batch)

                # Limpiar las listas para el siguiente lote
                image_paths = []
                shortcodes = []
                usernames = []

                # Liberar memoria
                del preds_batch
                gc.collect()

# Si hay imágenes restantes en las listas (último lote)
if image_paths:
    preds_batch = []
    for img_path in image_paths:
        pred = predict_image("vgg16", img_path)
        preds_batch.append(pred)

    predicciones['username'].extend(usernames)
    predicciones['post'].extend(shortcodes)
    predicciones['type'].extend(preds_batch)

    del preds_batch
    gc.collect()

# Crear un DataFrame con las predicciones
df_predicciones = pd.DataFrame(predicciones)

# Guardar el DataFrame en un nuevo archivo CSV
df_predicciones.to_csv(nuevo_csv_path, index=False)

```
